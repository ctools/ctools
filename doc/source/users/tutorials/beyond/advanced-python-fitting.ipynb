# Advanced model manipulation and fitting in Python

In this tutorial you will learn to perform some more advanced model manipulation and fitting from Python. You can also use them in a Python pipeline to test multiple model hypotheses and select the one that is most appropriate.

As usual we start by importing the gammalib, ctools, and cscripts Python modules.

```python
import gammalib
import ctools
import cscripts
```

For this tutorial we will also use the matplotlib package for plotting.

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

## Simulated dataset

For the tutorial we will simulate a small dataset. We start by defining the Instrument Response Function and energy range that we will use all along.

```python
caldb = 'prod2'
irf = 'South_0.5h'
emin = 0.1 #TeV
emax = 160. #TeV
```

Now we proceed to simulate the dataset. It consists of an hour of observations of the Crab nebula region, as usual pointed at a slightly offset position from the target. The input model is different from the one you have been using so far and contains some surprises. Don't look at it until you have completed the exercises at the end of the tutorial.

```python
evfile = 'events.fits'

obssim = ctools.ctobssim()

obssim['ra'] = 83.63
obssim['dec'] = 22.51
obssim['rad'] = 5.0
obssim['tmin'] = 0
obssim['tmax'] = 3600
obssim['emin'] = emin
obssim['emax'] = emax
obssim['caldb'] = caldb
obssim['irf'] = irf
obssim['inmodel'] = '$CTOOLS/share/models/crab_beyond.xml'
obssim['outevents'] = evfile

obssim.execute()
```

We have saved the events on disk in the file `events.fits`. In this way you can easily re-use the code below by substituting `events.fits` with your own event list of observation definition XML file.

We will peek at the simulated dataset by displaying a skymap.

```python
mapfile = 'skymap.fits'

skymap = ctools.ctskymap()

skymap['inobs'] = evfile
skymap['emin'] = emin
skymap['emax'] = emax
skymap['nxpix'] = 40
skymap['nypix'] = 40
skymap['binsz'] = 0.02
skymap['proj'] = 'TAN'
skymap['coordsys'] = 'CEL'
skymap['xref'] = 83.63
skymap['yref'] = 22.01
skymap['bkgsubtract'] = 'IRF'
skymap['caldb'] = caldb
skymap['irf'] = irf
skymap['outmap'] = mapfile

skymap.run()
```

```python
# We slightly smooth the map for display
skymap.skymap().smooth('GAUSSIAN',0.02)

from matplotlib.colors import SymLogNorm
# The SymLogNorm scale is a Log scale for both positive and negative values
# and is linear within a certain range around 0

ax = plt.subplot()
plt.imshow(skymap.skymap().array(),origin='lower',
           extent=[83.63+0.02*20,83.63-0.02*20,22.01-0.02*20,22.01+0.02*20], # boundaries of the coord grid
           norm=SymLogNorm(1)) # the scale will be linear within +-1 count 
ax.set_xlabel('R.A. (deg)')
ax.set_ylabel('Dec (deg)')
cbar = plt.colorbar()
cbar.set_label('Counts')
```

## Model fitting and residual inspection

First, we will fit the usual Crab model to the data. We will be using a 3D unbinned analysis.

```python
like = ctools.ctlike()

like['inobs'] = evfile
like['caldb'] = caldb
like['irf'] = irf
like['inmodel'] = '$CTOOLS/share/models/crab.xml'

like.run()
```

As usual it is wise to look at the output from the optimizer. We will also store the best-fit minus log-likelihood value for later usage.

```python
print(like.opt())

like1 = like.opt().value()
```

The fit has properly converged after 10 iterations. Let's take a look at the best-fit model.

```python
print(like.obs().models())
```

The background model reproduces well the data, since its best-fit normalization is 1 and the index/tilt is 0. The Crab nebula has a best-fit spectral index of 2.28.

To assess whether the model reproduces well the data spatially we will calculate a residual map.

```python
resmap = cscripts.csresmap(like.obs())

resmap['algorithm'] = 'SIGNIFICANCE'
resmap['emin'] = emin
resmap['emax'] = emax
resmap['nxpix'] = 40
resmap['nypix'] = 40
resmap['binsz'] = 0.02
resmap['proj'] = 'TAN'
resmap['coordsys'] = 'CEL'
resmap['xref'] = 83.63
resmap['yref'] = 22.01
resmap['outmap'] = 'resmap1.fits'

resmap.run()
```

We will be inspecting several of these residual maps along the tutorial, so we define a function to plot them, and apply it to our latest run.

```python
def plot_residuals(resid):
    #smoothing
    resid.smooth('GAUSSIAN',0.04)
    #plotting
    fig = plt.figure()
    ax = plt.subplot()
    plt.imshow(resid.array(),origin='lower',
               cmap='bwr',vmin=-3,vmax=3,
               extent=[83.63+0.02*20,83.63-0.02*20,22.01-0.02*20,22.01+0.02*20]) # boundaries of the coord grid
    ax.set_xlabel('R.A. (deg)')
    ax.set_ylabel('Dec (deg)')
    cbar = plt.colorbar()
    cbar.set_label('Significance ($\sigma$)')
```

```python
plot_residuals(resmap._resmap)
```

## Adding model components

The residual map shows a cluster of positive residuals on the South-West of the Crab nebula. Is this another source? To test this hypothesis we will add a point source at the position that we can guess from the residual map.

```python
newpntsrc = gammalib.GModelSky(gammalib.GModelSpatialPointSource(83.7,21.9),
                              gammalib.GModelSpectralPlaw(1.e-17,-2.,gammalib.GEnergy(3.e5,'MeV')),
                              gammalib.GModelTemporalConst(1))
```

We have defined a sky model object that has three components:
- a spatial model, which is a point source at the position guessed from the residual map;
- a spectral model which is a power law with spectral index 2, and a flux which approximately 1/10 of the Crab nebula;
- a temporal model which is a constant.
We will name this source Src1, and free its position (fixed by default when a new source is created), so that we can fit it to the data.

```python
newpntsrc.name('Src1')
newpntsrc['RA'].free()
newpntsrc['DEC'].free()
```

Finally we append the new source to the model

```python
like.obs().models().append(newpntsrc)
```

and we fit the model including the new source to the data.

```python
like.run()
```

Does the addition of the new source provide a better fit to the data? We can quantify this using the test statistic (TS) given by twice the log-likelihood difference.

```python
like2 = like.opt().value()

ts = -2 * (like2 - like1)
print(ts)
```

TS is expected to be distributed as a $\chi^2$ with n degrees of freedom, where n is the additional number of degrees of freedom in the model including the new source, in our case 4 (RA, Dec, Prefactor, and Index). The integral of the $\chi^2$ from TS to $\infty$ is the chance probability that the likelihood improved by that much due to statistical fluctuations. A large value, like the one we got, means that the chance probability is very low, thus we are likely to have found a new source.

To make sure the new source improves the data/model agreement we will look again at the residual map.

```python
resmap.models(like.obs().models())
resmap.run()

plot_residuals(resmap._resmap)
```

The addition of the new point source has flattened the spatial residuals, even though the fit is not perfect yet obvisouly. From now on we will fix the position of Scr1.

```python
like.obs().models()['Src1']['RA'].fix()
like.obs().models()['Src1']['DEC'].fix()
```

## Modifying the spectral models

Next we will check if the model agrees well with the data at all energies. We will start by defining a function to make and plot residual maps in separate energy ranges.

```python
def resmap_erange(ebounds):
    for elow, eup in ebounds:
        resmap['emin'] = elow
        resmap['emax'] = eup
        resmap.models(like.obs().models()) # we make sure we are using the latest best-fit model
        resmap.run()
        plot_residuals(resmap._resmap)
```

```python
resmap_erange([[emin,1.],
              [1.,10.],
              [10.,emax]])
```

The most obvious problem is a negative residual cluster at the position of the Crab nebula at the highest energies. We will change the spectral model of the crab nebula and test if this improves the fit. We will use an exponentially cutoff power-law rather than a simple power law. We will use as starting values for the Prefactor and Index the current best-fit values and tentatively set the cutoff energy at 1 TeV.

```python
crab = like.obs().models()['Crab']

expplaw = gammalib.GModelSpectralExpPlaw()

expplaw['Prefactor'].value(crab['Prefactor'].value())
expplaw['Index'].value(crab['Index'].value())
expplaw['PivotEnergy'].value(crab['PivotEnergy'].value())
expplaw['CutoffEnergy'].value(1.e6)

crab.spectral(expplaw)
```

We will now fit the modified model to the data

```python
like.run()
```

and use TS to quantify the model improvement (in this case we have only one additional degree of freedom, the ctuoff energy).

```python
like3 = like.opt().value()

ts = -2 * (like3 - like2)
print(ts)
```

The model improvement was again significant. Let's look at the fit results for the Crab nebula.

```python
print(crab)
```

The best-fit cutoff energy is 3.3 TeV, which seems consistent with the behavior in the energy-bands residual maps above. Note that this has changed significantly the Index value, that now is 1.83. We will check the residual maps again to make sure.

```python
resmap_erange([[emin,1.],
              [1.,10.],
              [10.,emax]])
```

Now we actually have an excess at the highest energies. In spite of the significant likelihood improvement the model is not quite perfect yet. Is there perhaps a new spectral component? We will test if the addition of a power-law spectral component. It will be added to the exponentially-cutoff power law in a composite spectral model.

```python
expplaw2 = like.obs().models()['Crab'].spectral().clone() 

newcomp = gammalib.GModelSpectralPlaw(1.e-18,-2.,gammalib.GEnergy(1,'TeV'))

comp_spec = gammalib.GModelSpectralComposite()
comp_spec.append(expplaw2)
comp_spec.append(newcomp)

like.obs().models()['Crab'].spectral(comp_spec)
```

Let's fit this new model to the data and check the likelihood improvement.

```python
like.run()

like4 = like.opt().value()
ts = -2 * (like4 - like3)
print(ts)
```

The likelihood improvement is again significant, but did we already find the best possible model of the sky?
